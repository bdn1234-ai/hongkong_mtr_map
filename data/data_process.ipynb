```python
import pandas as pd
```

```python
ds = pd.read_csv("MTR_Edges_Data.csv")
```

```python
from collections import defaultdict

graph = defaultdict(set)

for _, row in ds.iterrows():
    u = row["start_eng"].strip()
    v = row["end_eng"].strip()
    w = float(row["edge_distance_km"])

    graph[u].add((v, w))
    graph[v].add((u, w))
    
graph = {k: list(v) for k, v in graph.items()}
```

```python
df = pd.read_csv("/Users/admin/hongkong_mtr_map/data/Stations_With_Coords_And_Maps.csv")

df = df.drop_duplicates(subset=["station_eng"])

coords = {
    row["station_eng"]: [row["lat"], row["long"]]
    for _, row in df.iterrows()
}
```

```python
import json

with open("coords.json", "w", encoding="utf-8") as f:
    json.dump(coords, f, ensure_ascii=False, indent=2)

with open("graph.json", "w", encoding="utf-8") as f:
    json.dump(graph, f, ensure_ascii=False, indent=2)
```
